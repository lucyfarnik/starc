```python
import os
import json
import pandas as pd
```

```python
# loop over all files in temp/interpolated
temp_dir = '../temp/interpolated'
data = []
for filename in os.listdir(temp_dir):
  if filename == 'log.txt': continue
  # load json file
  with open(f'{temp_dir}/{filename}', 'r') as f:
    sample_results = json.load(f)
  # loop over all results in json file and add to data
  for result in sample_results:
    data.append(result)

# convert to pandas dataframe
df = pd.DataFrame(data)
df
```

```python
# show null rows
df[df.isnull().any(axis=1)]
```

```python
# average regret for each row
df['regret'] = (df['regret1'] + df['regret2']) / 2
```

```python
# rename columns to match the new naming convention
new_names = {}
for col in df.columns:
  if 'StateVal-' in col:
    new_names[col] = col.replace('StateVal-', 'VALPotential-')
  if 'StateValExp-' in col:
    new_names[col] = col.replace('StateValExp-', 'VAL-')
  if 'Minimal-' in col:
    new_names[col] = col.replace('Minimal-', 'MinimalPotential-')

df = df.rename(columns=new_names)
```

```python
# no_baselines = [col for col in df.columns if 'None' not in col and '0' not in col]
# corrs = df[no_baselines].corr()['regret'].sort_values(ascending=False)
corrs = df.corr()['regret'].sort_values(ascending=False)
print(corrs.to_string())
```

```python
string = corrs.to_string()
vals = string.split('\n')
vals = { val.split(' ')[0]: float(val.split(' ')[-1]) for val in vals }
vals
```

```python
for key, val in vals.items():
  if '-2-' in key:
    print(key, val)
```

```python
vals['EPIC-2-2'], vals['DARD-2-2'], vals['VAL-2-2']
```

```python
latex_table_format = ''
for key, val in vals.items():
  if 'regret' in key: continue
  # if "_" is in the key, escape it
  if '_' in key: key = key.replace('_', '\_')
  latex_table_format += f'{key} & {val:.3f} \\\\\n'
print(latex_table_format)
```

```python
canons = ['None', 'EPIC', 'DARD', 'MinimalPotential', 'VALPotential', 'VAL']
col_by_canon = { # columns organized by Canon used
    can: [col for col in df.columns if col.startswith(can+'-')]
    for can in canons
}
cols_by_canon_norm = { # only columns that are normalized
    can: [col for col in col_by_canon[can] if '0' not in col]
    for can in canons
}
cols_by_canon_no_norm = { # only non-normalized columns
    can: [col for col in col_by_canon[can] if '0' in col]
    for can in canons
}

corrs_by_canon_norm = { # correlations by canon, only normalized
    can: [vals[k] for k in cols_by_canon_norm[can]]
    for can in canons
}
corrs_by_canon_no_norm = { # correlations by canon, only non-normalized
    can: [vals[k] for k in cols_by_canon_no_norm[can]]
    for can in canons
}
(sum(corrs_by_canon_norm['EPIC'])/len(corrs_by_canon_norm['EPIC']),
 sum(corrs_by_canon_norm['VAL'])/len(corrs_by_canon_norm['VAL']))
```

```python
# visualize the two sets on a scatter plot (note that they aren't the same size, we want to show "is_s_prime" vs "is_potential" on one axis and the value on the other axis)
import matplotlib.pyplot as plt
markers = ['o', 's', '^', 'v', 'D', 'P']
colors = ['gray', 'darkgreen', 'deepskyblue', 'orange', 'navy', 'purple']
for i, can in enumerate(canons):
  label = f'STARC-{can}' if can == 'VAL' else can
  plt.scatter([i for _ in corrs_by_canon_norm[can]], corrs_by_canon_norm[can],
              label=label, marker=markers[i], color=colors[i])
  plt.scatter([i for _ in corrs_by_canon_no_norm[can]], corrs_by_canon_no_norm[can],
              marker=markers[i], color='gray')

# highlight EPIC-2-2 and DARD-2-2
plt.scatter([1], [vals['EPIC-2-2']], marker='s', color='red')
plt.scatter([2], [vals['DARD-2-2']], marker='^', color='red')

# style the plot
plt.gca().axes.get_xaxis().set_visible(False)
plt.ylabel('Correlation with regret')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# save as vector image
plt.savefig('../results/final/correlations.pdf', bbox_inches='tight')
plt.show()
```

```python
# check if for any normalisation or distance, VAL is better than EPIC
counter_true = 0
counter_false = 0
for key, val in vals.items():
  if 'EPIC' in key:
    val_key = key.replace('EPIC', 'VAL')
    if val < vals[val_key]:
      counter_true += 1
    else:
      counter_false += 1
      print(f"Exception: {key:<35} EPIC value={val:<15} VAL value={vals[val_key]}")
print(counter_true, counter_false)
```
