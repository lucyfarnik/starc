```python
import pandas as pd
import json

slippery = False

with open('../results/discount_changes.json', 'r') as f:
  r = json.load(f)
  results = r['results']
  config = r['config']

# compute the correlations to regret for each discount value
corr_results = []
for env_i, env_results in enumerate(results):
  env_res_flat = [result for reward_results in env_results for result in reward_results]
  for r in env_res_flat:
    r['regret'] = (r['regret1'] + r['regret2']) / 2
  df = pd.DataFrame(env_res_flat)
  corr = df.corr()['regret'].to_dict()
  corr['discount'] = config['low_discount'] * (config['high_discount'] / config['low_discount']) ** (env_i / (config['num_envs'] - 1))
  corr_results.append(corr)

# convert to dataframe
df = pd.DataFrame(corr_results)

df.head()
```
