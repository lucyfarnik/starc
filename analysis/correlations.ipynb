```python
import pandas as pd
import matplotlib.pyplot as plt
import json

results = None
with open('../results/5_consistent_rollouts.json', 'r') as f:
  results = json.load(f)
results = [r for env_rs in results for r_rs in env_rs for r in r_rs]
df = pd.DataFrame(results)
```

```python
df.corr()
```

```python
corr = df.corr().to_numpy()
corr.min(), corr.mean()
```

```python
plt.matshow(df.corr())
```

```python
df_no_regret = df.drop(['regret1', 'regret2'], axis=1)
df_no_regret.head()
```

```python
df_no_regret.corr()
```

```python
corr = df_no_regret.corr().to_numpy()
corr.min(), corr.mean()
```
