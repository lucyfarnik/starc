```python
import pandas as pd
import matplotlib.pyplot as plt
import json

results = None
with open('../results/4_minimal_diverse_r.json', 'r') as f:
  results = json.load(f)
results = [r for env_rs in results for r_rs in env_rs for r in r_rs]
df = pd.DataFrame(results)
```

```python
df.corr()
```

```python
corr = df.corr().to_numpy()
corr.min(), corr.mean()
```

```python
plt.matshow(df.corr())
```
